```python
# Q1 Extracting Tesla Stock Data Using yfinance

import yfinance as yf

# Define the stock symbol (Ticker) for Tesla
tesla_ticker = "TSLA"

# Define the date range for the data
start_date = "2020-01-01"
end_date = "2023-01-01"

# Use yfinance to fetch Tesla stock data
tesla_data = yf.download(tesla_ticker, start=start_date, end=end_date)

# Display the first few rows of the data
print(tesla_data.head())
```

```python
#Q2 Extracting Tesla Revenue Data Using Webscraping

import requests
from bs4 import BeautifulSoup

# URL of the website containing Tesla's revenue data
tesla_revenue_url = "https://example.com/tesla-revenue"

# Send a GET request to the website
response = requests.get(tesla_revenue_url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find the relevant HTML element containing the revenue data
revenue_element = soup.find("div", class_="revenue-data")

# Extract the revenue value
revenue = revenue_element.text

# Print the extracted revenue
print("Tesla's Revenue:", revenue)
```

```python
#Q5 Tesla Stock and Revenue Dashboard 

import yfinance as yf
import matplotlib.pyplot as plt

# Define the stock symbol (Ticker) for Tesla
tesla_ticker = "TSLA"

# Define the date range for the data
start_date = "2020-01-01"
end_date = "2023-01-01"

# Use yfinance to fetch Tesla stock data
tesla_data = yf.download(tesla_ticker, start=start_date, end=end_date)

# Sample revenue data (replace with actual scraped revenue)
tesla_revenue = [100000000, 150000000, 200000000, 250000000, 300000000]

# Extract dates and stock prices from Tesla data
dates = tesla_data.index
stock_prices = tesla_data['Adj Close']

# Create a figure and plot stock price
plt.figure(figsize=(10, 6))
plt.plot(dates, stock_prices, label='Stock Price', color='blue')

# Add revenue data as a bar plot
plt.bar(dates, tesla_revenue, label='Revenue', color='green', alpha=0.5)

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Tesla Stock Price vs Revenue')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
